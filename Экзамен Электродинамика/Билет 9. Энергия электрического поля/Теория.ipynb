**<div align="center">Электрическая энергия заряженного конденсатора</div>**

**Определение** $:$ электрическая энергия заряженного конденсатора — это энергия, накопленная в электрическом поле между его обкладками.

$$
W = \frac{1}{2}CU^2
$$

**Где** $:$

$W$ — энергия конденсатора $($Дж$)$  
$C$ — ёмкость конденсатора $($Ф$)$  
$U$ — напряжение между обкладками $($В$)$  

**Особенности** $:$
1. Энергия пропорциональна квадрату напряжения.
2. Энергия зависит от ёмкости конденсатора.


**<div align="center">Электрическая энергия заряженного проводника и системы заряженных проводников</div>**

**Определение** $:$ энергия заряженного проводника — это энергия, накопленная в его электрическом поле.

$$
W = \frac{1}{2}Q\varphi
$$

**Где** $:$

$W$ — энергия проводника $($Дж$)$  
$Q$ — заряд проводника $($Кл$)$  
$\varphi$ — потенциал проводника $($В$)$  

**Для системы проводников** $:$ энергия определяется как сумма энергий взаимодействия зарядов:

$$
W = \frac{1}{2} \sum_{i=1} Q_i \varphi_i
$$

**Особенности** $:$
1. Энергия системы проводников учитывает взаимодействие зарядов.
2. Энергия зависит от распределения зарядов и потенциалов.


**<div align="center">Энергия взаимодействия точечных зарядов</div>**

**Определение** $:$ энергия взаимодействия точечных зарядов — это энергия, накопленная в электрическом поле между зарядами.

$$
W = \frac{1}{4\pi\varepsilon_0}\sum_{i<j}\frac{q_i q_j}{r_{ ij}}
$$

**Где** $:$

$W$ — энергия взаимодействия $($Дж$)$  
$q_i, q_j$ — величины зарядов $($Кл$)$  
$r_{ij}$ — расстояние между зарядами $($м$)$  

**Особенности** $:$
1. Энергия взаимодействия уменьшается с увеличением расстояния между зарядами.
2. Энергия положительна для одноимённых зарядов и отрицательна для разноимённых.


**<div align="center">Плотность энергии электрического поля. Локализация энергии в пространстве</div>**

**Определение** $:$ плотность энергии электрического поля — это энергия, сосредоточенная в единице объёма пространства, заполненного электрическим полем.

$$
w = \frac{\varepsilon_0 \varepsilon E^2}{2}
$$

**Где** $:$

$w$ — плотность энергии электрического поля $($Дж/м³$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\varepsilon$ — диэлектрическая проницаемость материала $($безразмерная$)$  
$E$ — напряжённость электрического поля $($Н/Кл$)$  

**Локализация энергии** $:$ энергия электрического поля локализована в пространстве, где существует поле. Для расчёта полной энергии используется интеграл:

$$
W = \int\limits_V w \cdot dV = \int\limits_V \frac{\varepsilon_0 \varepsilon E^2}{2} \cdot dV
$$

**Особенности** $:$
1. Плотность энергии пропорциональна квадрату напряжённости поля.
2. Полная энергия поля определяется интегрированием плотности по объёму.
