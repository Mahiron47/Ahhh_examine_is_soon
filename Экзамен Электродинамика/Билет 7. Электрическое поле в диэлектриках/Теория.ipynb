**<div align="center">Поляризация диэлектриков</div>**

**Определение** $:$ поляризация диэлектриков — это процесс смещения зарядов внутри диэлектрика под действием внешнего электрического поля, приводящий к образованию связанных зарядов на его поверхности.

**Особенности** $:$
1. Внешнее поле вызывает смещение положительных и отрицательных зарядов внутри молекул диэлектрика.
2. На поверхности диэлектрика появляются связанные заряды, создающие собственное электрическое поле.
3. Поляризация приводит к уменьшению напряжённости внешнего поля внутри диэлектрика.


**<div align="center">Связанные и сторонние заряды</div>**

**Связанные заряды** $:$ заряды, возникающие на поверхности диэлектрика в результате его поляризации. Они обусловлены смещением зарядов внутри молекул.

**Сторонние заряды** $:$ заряды, отдельных ионов, те же, что и в проводниках.

**Особенности** $:$
Связанные заряды создают поле, противоположное внешнему.


**<div align="center">Вектор поляризации</div>**

**Определение** $:$ вектор поляризации $\vec{P}$ — это векторная величина, равная дипольному моменту единицы объёма диэлектрика.

$$
\vec{P} = \frac{\sum \vec{p}}{\Delta V}
$$

**Где** $:$

$\vec{P}$ — вектор поляризации $($Кл/м²$)$  
$\vec{p}$ — дипольный момент молекулы $($Кл·м$)$  
$V$ — объём диэлектрика $($м³$)$  

**Особенности** $:$
1. Направление вектора поляризации совпадает с направлением дипольных моментов молекул.
2. Поляризация зависит от напряжённости внешнего поля и свойств диэлектрика.


**<div align="center">Теорема Гаусса для вектора поляризации</div>**

**Формулировка** $:$ поток вектора поляризации через замкнутую поверхность равен величине связанных зарядов внутри этой поверхности.

$$
\oint\limits_S \vec{P} \cdot d\vec{S} = -q_{\text{связ}}
$$

**Где** $:$

$\vec{P}$ — вектор поляризации $($Кл/м²$)$  
$d\vec{S}$ — элементарная площадка поверхности $($м²$)$  
$q_{\text{связ}}$ — величина связанных зарядов $($Кл$)$  

**Особенности** $:$
1. Поток вектора поляризации отрицателен, так как связанные заряды противоположны сторонним.
2. Теорема используется для расчёта связанных зарядов.


**<div align="center">Вектор электрического смещения. Теорема Гаусса для вектора D</div>**

**Определение** $:$ вектор электрического смещения $\vec{D}$ — это векторная величина, характеризующая распределение сторонних зарядов в диэлектрике.

$$
\vec{D} = \varepsilon_0 \vec{E} + \vec{P}
$$

**Где** $:$

$\vec{D}$ — вектор электрического смещения $($Кл/м²$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$\vec{P}$ — вектор поляризации $($Кл/м²$)$  

**Теорема Гаусса для $\vec{D}$** $:$ поток вектора электрического смещения через замкнутую поверхность равен величине сторонних зарядов внутри этой поверхности.

$$
\oint\limits_S \vec{D} \cdot d\vec{S} = q_{\text{стор}}
$$

**Где** $:$

$q_{\text{стор}}$ — величина сторонних зарядов $($Кл$)$  

**Особенности** $:$
1. Поток вектора $\vec{D}$ положителен, так как он связан со сторонними зарядами.
2. Теорема используется для расчёта напряжённости поля в диэлектрике.


**<div align="center">Дифференциальные соотношения</div>**

**Для $\vec{P}$** $:$ дивергенция вектора поляризации равна плотности связанных зарядов:

$$
\nabla \cdot \vec{P} = -\rho_{\text{связ}}
$$

**Для $\vec{D}$** $:$ дивергенция вектора электрического смещения равна плотности сторонних зарядов:

$$
\nabla \cdot \vec{D} = \rho_{\text{стор}} 
$$

**Где** $:$

$\rho_{\text{связ}}$ — плотность связанных зарядов $($Кл/м³$)$  
$\rho_{\text{стор}}$ — плотность сторонних зарядов $($Кл/м³$)$  

**Особенности** $:$
1. Дифференциальные соотношения используются для локального анализа распределения зарядов.
2. Они связывают макроскопические параметры поля с зарядовой плотностью.


**<div align="center">Пример: Поле однородно заряженного диэлектрического шара</div>**

**Условие задачи** $:$ Однородно заряженный диэлектрический шар радиуса $R$ имеет объёмную плотность сторонних зарядов $\rho_{\text{стор}}$. Требуется найти напряжённость электрического поля внутри и вне шара.

**Решение** $:$

1. **Внутри шара** $:$

$$
\vec{D} = \frac{\rho_{\text{ст}} r}{3}, \quad \vec{E} = \frac{\rho_{\text{ст}} r}{3 \varepsilon\varepsilon_0}, \quad \vec{P} = \vec{D} - \varepsilon_0 \vec{E}.
$$

2. **Вне шара** $:$

$$
\vec{D} = \frac{Q}{4 \pi R^2}, \quad \vec{E} = \frac{Q}{4 \pi \varepsilon_0 R^2}, \quad P = 0
$$

**Где** $:$

$Q$ — полный заряд шара $($Кл$)$  
$r$ — расстояние от центра шара $($м$)$  
$\varepsilon$ — диэлектрическая проницаемость материала шара  

**Особенности** $:$
1. Внутри шара поле линейно возрастает с расстоянием от центра.
2. Вне шара поле убывает как $1/r^2$.
3. Поле зависит от диэлектрической проницаемости материала.
