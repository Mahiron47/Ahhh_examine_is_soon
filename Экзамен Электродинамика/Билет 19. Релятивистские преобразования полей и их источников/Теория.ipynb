**<div align="center">Инвариантность заряда</div>**

**Определение** $:$ электрический заряд является инвариантной величиной, то есть его значение не зависит от выбора системы отсчёта.

**Особенности** $:$
1. Заряд сохраняется при любых преобразованиях Лоренца.
2. Инвариантность заряда лежит в основе закона сохранения заряда.


**<div align="center">Законы преобразования полей E и B</div>**

**Формулы** $:$ при переходе между системами отсчёта электрическое и магнитное поля преобразуются следующим образом:

1. Электрическое поле:

$$
\vec{E}' = \gamma \left(\vec{E} + \vec{v} \times \vec{B}\right) - \frac{\gamma^2}{\gamma + 1} \vec{v} \left(\vec{v} \cdot \vec{E}\right)
$$

2. Магнитное поле:

$$
\vec{B}' = \gamma \left(\vec{B} - \frac{\vec{v} \times \vec{E}}{c^2}\right) - \frac{\gamma^2}{\gamma + 1} \vec{v} \left(\vec{v} \cdot \vec{B}\right)
$$

**Где** $:$

$\gamma$ — фактор Лоренца $\gamma = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}}$  
$\vec{v}$ — скорость системы отсчёта относительно другой $($м/с$)$  
$c$ — скорость света $($м/с$)$  

**Особенности** $:$
1. Электрическое и магнитное поля связаны и преобразуются друг в друга.
2. Преобразования зависят от скорости системы отсчёта.


**<div align="center">Релятивистская природа магнетизма</div>**

**Определение** $:$ магнетизм является релятивистским эффектом, возникающим из-за движения зарядов относительно наблюдателя.

**Особенности** $:$
1. Магнитное поле — это проявление электрического поля в движущейся системе отсчёта.
2. Магнитные силы зависят от скорости движения зарядов.
3. Магнитное поле связано с электрическим через преобразования Лоренца.


**<div align="center">Вывод формулы для магнитной силы на примере заряженного стержня</div>**

**Условие** $:$ рассмотрим стержень длиной $l$, равномерно заряженный с линейной плотностью заряда $\lambda$, движущийся со скоростью $v$.

**Решение** $:$
1. Магнитное поле, создаваемое движущимся зарядом:

$$
B = \frac{\mu_0 \lambda v}{2\pi r}
$$

2. Сила Лоренца, действующая на заряд $q$ в магнитном поле:

$$
F = qvB = qv \cdot \frac{\mu_0 \lambda v}{2\pi r}
$$

**Где** $:$

$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$r$ — расстояние от стержня до точки наблюдения $($м$)$  
$v$ — скорость стержня $($м/с$)$  
$q$ — заряд $($Кл$)$  

**Особенности** $:$
1. Магнитная сила пропорциональна квадрату скорости.
2. Сила зависит от расстояния до стержня.


**<div align="center">Инварианты электромагнитного поля</div>**

**Определение** $:$ инварианты электромагнитного поля — это величины, сохраняющие своё значение при преобразованиях Лоренца.

**Формулы** $:$
1. Первый инвариант:

$$
I_1 = \vec{B} \cdot \vec{E}
$$

2. Второй инвариант:

$$
I_2 = B^2 - \frac{E^2}{c^2}
$$

**Особенности** $:$
1. Инварианты используются для анализа электромагнитных полей.
2. Они помогают выявить свойства поля, независимые от системы отсчёта.


**<div align="center">Магнитное поле заряда, движущегося с нерелятивистской скоростью</div>**

**Формула** $:$ магнитное поле, создаваемое движущимся зарядом, определяется как:

$$
\vec{B} = \frac{\mu_0}{4\pi} \cdot \frac{q \cdot \vec{v} \times \vec{r}}{r^3}
$$

**Где** $:$

$\vec{B}$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$q$ — заряд $($Кл$)$  
$\vec{v}$ — скорость заряда $($м/с$)$  
$\vec{r}$ — радиус-вектор от заряда до точки наблюдения $($м$)$  
$r$ — расстояние от заряда до точки наблюдения $($м$)$  

**Особенности** $:$
1. Магнитное поле пропорционально скорости заряда.
2. Поле убывает как $1/r^2$ с увеличением расстояния.
