**<div align="center">Электрический диполь. Точечный диполь. Дипольный момент</div>**

**Определение** $:$ электрический диполь — это система из двух точечных зарядов, равных по величине и противоположных по знаку, расположенных на некотором расстоянии друг от друга.

**Точечный диполь** $:$ идеализированная модель диполя, в которой расстояние между зарядами стремится к нулю, а произведение заряда на расстояние остаётся постоянным.

**Дипольный момент** $:$ векторная величина, характеризующая электрический диполь и равная произведению заряда $q$ на вектор, соединяющий заряды:

$$
\vec{p} = q\vec{L}
$$

**Где** $:$

$\vec{p}$ — дипольный момент $($Кл·м$)$  
$q$ — величина заряда $($Кл$)$  
$\vec{L}$ — вектор, соединяющий заряды, направленный от отрицательного заряда к положительному $($м$)$  

**Особенности** $:$
1. Дипольный момент является характеристикой распределения зарядов в системе.
2. Направление дипольного момента совпадает с направлением от отрицательного заряда к положительному.


**<div align="center">Потенциал поля диполя</div>**

**Определение** $:$ потенциал электрического поля диполя в точке пространства определяется как сумма потенциалов, создаваемых каждым зарядом диполя.

$$
\varphi = \frac{1}{4\pi\varepsilon_0}\frac{\vec{p} \cdot \vec{r}}{r^3}
$$

**Где** $:$

$\varphi$ — потенциал поля диполя $($В$)$  
$\vec{p}$ — дипольный момент $($Кл·м$)$  
$\vec{r}$ — радиус-вектор, соединяющий центр диполя с точкой наблюдения $($м$)$  
$r$ — расстояние от центра диполя до точки наблюдения $($м$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  

**Особенности** $:$
1. Потенциал поля диполя убывает как $1/r^2$ с увеличением расстояния.
2. Потенциал диполя зависит от ориентации диполя относительно точки наблюдения.


**<div align="center">Напряжённость электрического поля диполя</div>**

**Определение** $:$ напряжённость электрического поля диполя определяется как

$$
\vec{E} = \frac{1}{4\pi\varepsilon_0} \cdot \frac{p}{r^3}\sqrt{1+3\cos^2{\theta}}
$$

**Где** $:$

$\vec{E}$ — напряжённость электрического поля диполя $($Н/Кл$)$  
$\vec{p}$ — дипольный момент $($Кл·м$)$  
$r$ — расстояние от центра диполя до точки наблюдения $($м$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\theta$ — угол между дипольным моментом и радиус вектором $($Рад$)$

**Особенности** $:$
1. Напряжённость поля диполя убывает как $1/r^3$ с увеличением расстояния.
2. Напряжённость поля зависит от ориентации диполя относительно точки наблюдения.


**<div align="center">Силы, действующие на диполь в электрическом поле</div>**

**Определение** $:$ на электрический диполь в неоднородном электрическом поле действуют силы и моменты, стремящиеся повернуть диполь вдоль линий напряжённости поля.

**Сила** $:$ результирующая сила, действующая на диполь, определяется как градиент энергии диполя в поле:

$$
F = \frac{1}{4\pi\varepsilon_0}\frac{2pQ}{r^3}
$$

**Момент сил** $:$ момент сил, действующий на диполь, определяется как векторное произведение дипольного момента и напряжённости поля:

$$
\vec{M} = \vec{p} \times \vec{E}
$$

**Где** $:$

$\vec{F}$ — сила, действующая на диполь $($Н$)$  
$\vec{M}$ — момент сил $($Н·м$)$  
$\vec{p}$ — дипольный момент $($Кл·м$)$  
$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  

**Особенности** $:$
1. В однородном поле результирующая сила на диполь равна нулю, но действует момент сил.
2. В неоднородном поле диполь перемещается в направлении увеличения напряжённости поля.
