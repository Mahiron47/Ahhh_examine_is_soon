**Электрический заряд** — скалярная величина, определяющая способность тел участвовать в электромагнитном взаимодействии

**Кулон** — это заряд, который проходит через поперечное сечение проводника за секунду при силе тока в ампер.
$$
1 \text{ Кл} = 1 \text{ А} \cdot 1 \text{ с}
$$

**<div align="center">Свойства электрического заряда</div>**

1.  **Взаимодействие** $:$ существует два вида зарядов — положительные и отрицательные. Одноимённые заряды отталкиваются, разноимённые — притягиваются.
2.  **Закон сохранения** $:$ полный заряд замкнутой системы постоянен.
3.  **Релятивистская инвариантность** $:$ величина заряда не зависит от системы отсчёта и его скорости.
4.  **Дискретность** $:$ заряд любого тела кратен элементарному заряду $e \approx 1.6 \cdot 10^{-19} \text{ Кл}$.
    *   Заряд электрона $:$ $-e$
    *   Заряд протона $:$ $+e$

---

**Пробный заряд** — это точечный заряд, который вносится в электростатическое поле для его исследования и удовлетворяет двум условиям:

1.  Его величина настолько мала, что он не вызывает заметного перераспределения исследуемых зарядов.
2.  Его размеры пренебрежимо малы по сравнению с расстояниями до зарядов, создающих поле.
