```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры эллипсоида
a = 4  # полуось по x
b = 3  # полуось по y

# Создаём массив углов
theta = np.linspace(0, 2 * np.pi, 100)

# Вычисляем координаты точек эллипса
x = a * np.cos(theta)
y = b * np.sin(theta)

# Находим фокусы эллипса
c = np.sqrt(a**2 - b**2)
f1 = (c, 0)
f2 = (-c, 0)

# Строим эллипс без интерфейса на прозрачном фоне
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y, color='white')  # цвет линии эллипса белый

# Подписываем точки пересечения с осями
# x = ±a, y = 0
ax.scatter([a, -a], [0, 0], color='white', zorder=5)
ax.text(a, 0.1, '$(a, 0)$', color='white', ha='left', va='bottom', fontsize=12)
ax.text(-a, 0.1, '$(-a, 0)$', color='white', ha='right', va='bottom', fontsize=12)
# x = 0, y = ±b
ax.scatter([0, 0], [b, -b], color='white', zorder=5)
ax.text(0.1, b, '$(0, b)$', color='white', ha='left', va='bottom', fontsize=12)
ax.text(0.1, -b, '$(0, -b)$', color='white', ha='left', va='top', fontsize=12)

# Обозначаем фокусы как точки
ax.scatter([f1[0], f2[0]], [f1[1], f2[1]], color='white', s=60, marker='o', zorder=6)
ax.text(f1[0]+0.1, f1[1], '$F_1$', color='white', ha='left', va='bottom', fontsize=12)
ax.text(f2[0]-0.1, f2[1], '$F_2$', color='white', ha='right', va='bottom', fontsize=12)

ax.set_aspect('equal')
ax.axis('off')
fig.patch.set_alpha(0.0)
ax.set_facecolor('none')
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
plt.show()
```
