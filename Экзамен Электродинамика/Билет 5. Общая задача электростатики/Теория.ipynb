**<div align="center">Теорема Гаусса в дифференциальной форме</div>**

**Формулировка** : дивергенция вектора напряжённости электрического поля равна отношению объёмной плотности заряда к электрической постоянной ε₀.

$$
\nabla \cdot \vec{E} = \frac{\rho}{\varepsilon_0}
$$

**Где** :  

$\vec{E}$ — напряжённость электрического поля (Н/Кл)  
$\rho$ — объёмная плотность заряда (Кл/м³)  
$\varepsilon_0$ — электрическая постоянная (Ф/м)  

**Особенности** :  
1. Теорема Гаусса в дифференциальной форме связывает локальные свойства электрического поля с распределением зарядов.  
2. Уравнение применимо для произвольных распределений зарядов.  

**<div align="center">Формулировка общей задачи электростатики</div>**

**Формулировка** : задача электростатики заключается в нахождении электрического поля $\vec{E}$ и потенциала $\varphi$ для заданного распределения зарядов $\rho$.

**Уравнения** :  
1. Уравнение Гаусса :  
$$
\nabla \cdot \vec{E} = \frac{\rho}{\varepsilon_0}
$$

2. Связь напряжённости с потенциалом :  
$$
\vec{E} = - \nabla \varphi
$$

**Где** :  

$\rho$ — объёмная плотность заряда (Кл/м³)  
$\varepsilon_0$ — электрическая постоянная (Ф/м)  
$\varphi$ — потенциал (В)  

**Особенности** :  
1. Задача электростатики является краевой задачей, требующей задания граничных условий.  
2. Решение задачи позволяет определить распределение поля и потенциала в пространстве.  

**<div align="center">Уравнение Пуассона</div>**

**Формулировка** : уравнение Пуассона связывает потенциал $\varphi$ с объёмной плотностью заряда $\rho$.

$$
\nabla\cdot\nabla\varphi = - \frac{\rho_{стор}}{\varepsilon\varepsilon_0}
$$

**Где** :  

$\varphi$ — потенциал (В)  
$\rho_{стор}$ — объёмная плотность стороннего заряда (Кл/м³)  
$\varepsilon_0$ — электрическая постоянная (Ф/м)  

**Особенности** :  
1. Уравнение Пуассона применяется для областей, содержащих заряды.  
2. Решение уравнения требует задания граничных условий.  

**<div align="center">Уравнение Лапласа</div>**

**Формулировка** : уравнение Лапласа описывает потенциал $\varphi$ в областях, свободных от зарядов.

$$
\nabla\cdot\nabla\varphi = 0
$$

**Где** :   

$\varphi$ — потенциал (В)  

**Особенности** :  
1. Уравнение Лапласа применяется для областей, где $\rho = 0$.  
2. Решение уравнения также требует задания граничных условий.  

**<div align="center">Теорема единственности</div>**

**Формулировка** : решение задачи электростатики для заданного распределения зарядов и граничных условий единственно.

**Особенности** :  
1. Теорема гарантирует, что найденное решение уравнений электростатики является единственным.  
2. Теорема единственности позволяет использовать различные методы для нахождения решения, зная, что результат будет одинаковым.  
3. Гарантия единственности решения важна для корректности физических расчётов.
