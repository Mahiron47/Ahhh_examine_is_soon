**<div align="center">Экспериментальные факты, лежащие в основе магнитных явлений</div>**

**Основные факты** $:$
1. Магнитное поле создаётся движущимися электрическими зарядами или токами.
2. Магнитное поле действует на движущиеся заряды, изменяя направление их движения.
3. Магнитное поле взаимодействует с другими магнитными полями, создавая силы и моменты.
4. Магнитное поле описывается векторной величиной $\vec{B}$, называемой магнитной индукцией.

**Особенности** $:$
1. Магнитное поле не действует на покоящиеся заряды.
2. Магнитное поле создаёт замкнутые силовые линии.


**<div align="center">Сила Лоренца</div>**

**Определение** $:$ сила Лоренца — это сила, действующая на движущийся заряд в магнитном и электрическом полях.

$$
\vec{F} = q\left(\vec{E} + \vec{v} \times \vec{B}\right)
$$

**Где** $:$

$\vec{F}$ — сила Лоренца $($Н$)$  
$q$ — заряд $($Кл$)$  
$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$\vec{v}$ — скорость заряда $($м/с$)$  
$\vec{B}$ — магнитная индукция $($Тл$)$  

**Особенности** $:$
1. Сила Лоренца перпендикулярна направлению движения заряда и магнитного поля.
2. Магнитная составляющая силы Лоренца не изменяет скорость заряда, а только его направление.


**<div align="center">Вывод формулы для силы Ампера</div>**

**Определение** $:$ сила Ампера — это сила, действующая на проводник с током в магнитном поле.

**Вывод** $:$
1. Рассмотрим элемент проводника длиной $dl$, по которому течёт ток $I$. На движущиеся заряды действует сила Лоренца:

$$
d\vec{F} = Id\vec{l} \times \vec{B}
$$

2. Для всего проводника сила Ампера равна интегралу:

$$
\vec{F} = \int Id\vec{l} \times \vec{B}
$$

**Где** $:$

$\vec{F}$ — сила Ампера $($Н$)$  
$I$ — сила тока $($А$)$  
$d\vec{l}$ — элемент длины проводника $($м$)$  
$\vec{B}$ — магнитная индукция $($Тл$)$  

**Особенности** $:$
1. Сила Ампера перпендикулярна проводнику и магнитному полю.
2. Сила зависит от длины проводника, силы тока и магнитной индукции.


**<div align="center">Закон Био-Савара</div>**

**Формулировка** $:$ закон Био-Савара описывает магнитное поле, создаваемое элементом проводника с током.

$$
\vec{B} = \oint\limits_L\frac{\mu_0}{4\pi} \frac{Id\vec{l} \times \vec{r}}{r^3}
$$

**Где** $:$

$d\vec{B}$ — элемент магнитной индукции $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$I$ — сила тока $($А$)$  
$d\vec{l}$ — элемент длины проводника $($м$)$  
$\vec{r}$ — радиус-вектор от элемента проводника до точки наблюдения $($м$)$  
$r$ — расстояние от элемента проводника до точки наблюдения $($м$)$  

**Особенности** $:$
1. Магнитное поле убывает как $1/r^2$ с увеличением расстояния.
2. Направление поля определяется правилом правого винта.


**<div align="center">Магнитное поле равномерно движущегося с нерелятивистской скоростью заряда</div>**

**Формула** $:$ магнитное поле, создаваемое движущимся зарядом, определяется как:

$$
\vec{B} = \frac{\mu_0}{4\pi}\frac{q\vec{v} \times \vec{r}}{r^3}
$$

**Где** $:$

$\vec{B}$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$q$ — заряд $($Кл$)$  
$\vec{v}$ — скорость заряда $($м/с$)$  
$\vec{r}$ — радиус-вектор от заряда до точки наблюдения $($м$)$  
$r$ — расстояние от заряда до точки наблюдения $($м$)$  

**Особенности** $:$
1. Магнитное поле зависит от скорости заряда и его направления.
2. Поле убывает как $1/r^2$ с увеличением расстояния.


**<div align="center">Сравнение сил магнитного и электрического взаимодействий движущихся зарядов</div>**

**Электрическая сила** $:$

$$
\vec{F}_E = q\vec{E}
$$

**Магнитная сила** $:$

$$
\vec{F}_B = q\vec{v} \times \vec{B}
$$

**Сравнение** $:$
1. Электрическая сила действует на заряд независимо от его движения.
2. Магнитная сила действует только на движущийся заряд.
3. Электрическая сила изменяет скорость заряда, а магнитная — только направление.
4. Магнитная сила пропорциональна скорости заряда.


**<div align="center">Магнитное поле прямого проводника с током</div>**

**Вывод формулы** $:$
1. Рассмотрим элемент проводника длиной $dl$, создающий магнитное поле в точке на расстоянии $r$.

$$
d\vec{B} = \frac{\mu_0}{4\pi}\frac{Id\vec{l} \times \vec{r}}{r^3}
$$

2. Интегрируем по всей длине проводника, используя принцип суперпозиции:

$$
\vec{B} = \frac{\mu_0 I}{2\pi r}
$$

**Где** $:$

$\vec{B}$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$I$ — сила тока $($А$)$  
$r$ — расстояние от проводника до точки наблюдения $($м$)$  

**Особенности** $:$
1. Магнитное поле убывает как $1/r$ с увеличением расстояния.
2. Направление поля определяется правилом правого винта.
