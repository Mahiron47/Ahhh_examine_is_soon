**<div align="center">Плотность тока. Сила тока</div>**

**Плотность тока** $:$ векторная величина, характеризующая количество заряда, проходящего через единичную площадку в единицу времени:

$$
j = \frac{\Delta q}{\Delta t \cdot \Delta S}
$$

**Сила тока** $:$ скалярная величина, равная количеству заряда, проходящего через проводник за единицу времени:

$$
I = \frac{\Delta q}{\Delta t}
$$

**Где** $:$

$\vec{j}$ — плотность тока $($А/м²$)$  
$I$ — сила тока $($А$)$  
$\Delta q$ — заряд $($Кл$)$  
$\Delta t$ — время $($с$)$  
$\Delta S$ — площадь поперечного сечения проводника $($м²$)$  

**Особенности** $:$
1. Направление плотности тока совпадает с направлением движения положительных зарядов.
2. Сила тока пропорциональна плотности тока и площади проводника.


**<div align="center">Уравнение непрерывности</div>**

**Формулировка** $:$ уравнение непрерывности выражает закон сохранения заряда в дифференциальной форме:

$$
\nabla \cdot \vec{j} + \frac{\partial \rho}{\partial t} = 0
$$

**Где** $:$

$\vec{j}$ — плотность тока $($А/м²$)$  
$\rho$ — плотность заряда $($Кл/м³$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Уравнение непрерывности связывает изменение плотности заряда с дивергенцией плотности тока.
2. Оно выражает локальный закон сохранения заряда.


**<div align="center">Закон Ома в локальной форме</div>**

**Формулировка** $:$ напряжённость электрического поля в проводнике пропорциональна плотности тока:

$$
\vec{E} = \rho \cdot \vec{j}
$$

**Где** $:$

$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$\rho$ — удельное сопротивление материала $($Ом·м$)$  
$\vec{j}$ — плотность тока $($А/м²$)$  

**Особенности** $:$
1. Закон Ома в локальной форме связывает макроскопические параметры проводника с микроскопическими характеристиками.
2. Удельное сопротивление зависит от материала проводника.


**<div align="center">Закон Джоуля-Ленца</div>**

**Формулировка** $:$ количество тепла, выделяемого в проводнике, пропорционально квадрату силы тока, сопротивлению проводника и времени:

$$
Q = I^2 R\Delta t
$$

**Где** $:$

$Q$ — количество тепла $($Дж$)$  
$I$ — сила тока $($А$)$  
$R$ — сопротивление проводника $($Ом$)$  
$\Delta t$ — время $($с$)$  

**Особенности** $:$
1. Закон Джоуля-Ленца описывает тепловые потери в проводнике.
2. Тепловые потери увеличиваются с ростом силы тока и сопротивления.


**<div align="center">Электродвижущая сила. Закон Ома для участка цепи, содержащего ЭДС</div>**

**Электродвижущая сила (ЭДС)** $:$ скалярная величина, равная работе сторонних сил по перемещению единичного заряда внутри источника:

$$
\mathcal{E} = \frac{A_{стор}}{\Delta q}
$$

**Закон Ома для участка цепи** $:$ сила тока в участке цепи равна отношению разности потенциалов к сопротивлению:

$$
I = \frac{\mathcal{E} - U}{R}
$$

**Где** $:$

$\mathcal{E}$ — электродвижущая сила $($В$)$  
$A$ — работа сторонних сил $($Дж$)$  
$q$ — заряд $($Кл$)$  
$I$ — сила тока $($А$)$  
$U$ — напряжение на участке цепи $($В$)$  
$R$ — сопротивление участка цепи $($Ом$)$  

**Особенности** $:$
1. ЭДС характеризует источник энергии в цепи.
2. Закон Ома учитывает влияние источника на ток в цепи.


**<div align="center">Закон Ома для замкнутой цепи</div>**

**Формулировка** $:$ сила тока в замкнутой цепи равна отношению ЭДС источника к сумме сопротивлений внешнего и внутреннего участка цепи:

$$
I = \frac{\mathcal{E}}{R + r}
$$

**Где** $:$

$I$ — сила тока $($А$)$  
$\mathcal{E}$ — электродвижущая сила $($В$)$  
$R$ — сопротивление внешнего участка цепи $($Ом$)$  
$r$ — внутреннее сопротивление источника $($Ом$)$  

**Особенности** $:$
1. Закон Ома для замкнутой цепи учитывает внутреннее сопротивление источника.
2. Сила тока уменьшается с увеличением сопротивления внешнего участка.


**<div align="center">Правила Кирхгофа</div>**

**Первое правило** $:$ сумма токов, входящих в узел, равна сумме токов, выходящих из узла:

$$
\sum I_{\text{вх}} = \sum I_{\text{вых}}
$$

**Второе правило** $:$ сумма ЭДС в замкнутом контуре равна сумме падений напряжения на всех элементах контура:

$$
\sum \mathcal{E} = \sum IR
$$

**Особенности** $:$
1. Правила Кирхгофа используются для расчёта сложных электрических цепей.
2. Первое правило выражает закон сохранения заряда.
3. Второе правило выражает закон сохранения энергии.


**<div align="center">Процессы при зарядке и разрядке конденсатора</div>**

**Зарядка конденсатора** $:$ процесс накопления заряда на обкладках конденсатора при подключении к источнику напряжения.

$$
q(t) = Q \cdot \left(1 - e^{-\frac{t}{RC}}\right)
$$

**Разрядка конденсатора** $:$ процесс освобождения заряда с обкладок конденсатора при подключении к резистору.

$$
q(t) = Q \cdot e^{-\frac{t}{RC}}
$$

**Где** $:$

$q(t)$ — заряд на обкладках конденсатора в момент времени $t$ $($Кл$)$  
$Q$ — максимальный заряд $($Кл$)$  
$R$ — сопротивление цепи $($Ом$)$  
$C$ — ёмкость конденсатора $($Ф$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Зарядка и разрядка происходят по экспоненциальному закону.
2. Постоянная времени $\tau = RC$ определяет скорость процессов.
