**<div align="center">Линии поля B</div>**

**Определение** $:$ линии поля $\vec{B}$ — это воображаемые линии, касательные к которым в каждой точке совпадают с направлением вектора магнитной индукции.

**Особенности** $:$
1. Линии поля $\vec{B}$ замкнуты, они не имеют начала и конца.
2. Направление линий поля определяется правилом правого винта.
3. Плотность линий поля пропорциональна величине магнитной индукции.

**Примеры** $:$
1. Для прямого проводника с током линии поля имеют форму концентрических окружностей.
2. Внутри соленоида линии поля параллельны и равномерно распределены.


**<div align="center">Теорема Гаусса для вектора B</div>**

**Формулировка** $:$ поток вектора магнитной индукции через любую замкнутую поверхность равен нулю:

$$
\oint\limits_S \vec{B} \cdot d\vec{S} = 0
$$

**Где** $:$

$\vec{B}$ — магнитная индукция $($Тл$)$  
$d\vec{S}$ — элементарная площадка поверхности $($м²$)$  

**Особенности** $:$
1. Теорема Гаусса для $\vec{B}$ отражает отсутствие магнитных зарядов.
2. Линии магнитного поля всегда замкнуты.


**<div align="center">Теорема о циркуляции вектора магнитной индукции</div>**

**Формулировка** $:$ циркуляция вектора магнитной индукции вдоль замкнутого контура равна произведению магнитной постоянной $\mu_0$ на алгебраическую сумму токов, охватываемых этим контуром:

$$
\oint\limits_L \vec{B} \cdot d\vec{l} = \mu_0I_{\text{внутр}}
$$

**Где** $:$

$\vec{B}$ — магнитная индукция $($Тл$)$  
$d\vec{l}$ — элемент длины контура $($м$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$I_{\text{внутр}}$ — сумма токов внутри контура $($А$)$  

**Особенности** $:$
1. Теорема используется для расчёта магнитного поля токов.
2. Направление обхода контура определяется правилом правого винта.


**<div align="center">Магнитное поле прямого бесконечно длинного провода с током</div>**

**Формула** $:$ магнитное поле вокруг бесконечно длинного прямого проводника с током определяется как:

$$
B = \frac{\mu_0 I}{2\pi r}
$$

**Где** $:$

$B$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$I$ — сила тока $($А$)$  
$r$ — расстояние от проводника до точки наблюдения $($м$)$  

**Особенности** $:$
1. Магнитное поле убывает как $1/r$ с увеличением расстояния.
2. Направление поля определяется правилом правого винта.


**<div align="center">Магнитное поле тороидальной катушки</div>**

**Формула** $:$ магнитное поле внутри тороидальной катушки определяется как:

$$
B = \frac{\mu_0 N I}{2\pi R}
$$

**Где** $:$

$B$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$N$ — число витков катушки  
$I$ — сила тока $($А$)$  
$R$ — радиус тороида $($м$)$  

**Особенности** $:$
1. Магнитное поле внутри тороида однородно.
2. Вне тороида магнитное поле практически отсутствует.


**<div align="center">Магнитное поле длинного соленоида</div>**

**Формула** $:$ магнитное поле внутри длинного соленоида определяется как:

$$
B = \mu_0 n I
$$

**Где** $:$

$B$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$n$ — число витков на единицу длины соленоида $($м⁻¹$)$  
$I$ — сила тока $($А$)$  

**Особенности** $:$
1. Магнитное поле внутри соленоида однородно и направлено вдоль его оси.
2. Вне соленоида магнитное поле практически отсутствует.
