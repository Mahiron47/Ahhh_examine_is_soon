**<div align="center">Вектор намагниченности</div>**

**Определение** $:$ вектор намагниченности $\vec{J}$ — это векторная величина, характеризующая степень намагниченности материала и равная магнитному моменту единицы объёма.

$$
\vec{J} = \frac{\sum \vec{p_m}}{\Delta V}
$$

**Где** $:$

$\vec{J}$ — вектор намагниченности $($А/м$)$  
$\vec{m}$ — магнитный момент молекулы $($А·м²$)$  
$V$ — объём материала $($м³$)$  

**Особенности** $:$
1. Направление вектора намагниченности совпадает с направлением магнитных моментов молекул.
2. Намагниченность зависит от внешнего магнитного поля и свойств материала.


**<div align="center">Теорема о циркуляции вектора намагниченности</div>**

**Формулировка** $:$ циркуляция вектора намагниченности вдоль замкнутого контура равна сумме связанных токов, охватываемых этим контуром:

$$
\oint\limits_L \vec{J} \cdot d\vec{l} = I'
$$

**Где** $:$

$\vec{J}$ — вектор намагниченности $($А/м$)$  
$d\vec{l}$ — элемент длины контура $($м$)$  
$I'$ — ток намагниченности $($А$)$  

**Особенности** $:$
1. Теорема используется для расчёта связанных токов.
2. Направление обхода контура определяется правилом правого винта.


**<div align="center">Магнитная восприимчивость. Магнитная проницаемость. Вектор $\vec{H}$</div>**

**Магнитная восприимчивость** $:$ безразмерная величина, характеризующая способность материала к намагничиванию:

$$
\chi_m = \frac{J}{H}
$$

**Магнитная проницаемость** $:$ величина, характеризующая способность материала проводить магнитное поле:

$$
\mu = 1 + \chi_m
$$

**Вектор $\vec{H}$** $:$ вектор напряжённости магнитного поля, связанный с вектором магнитной индукции и намагниченности:

$$
\vec{H} = \frac{\vec{B}}{\mu_0} - \vec{J}
$$

**Где** $:$

$\vec{H}$ — напряжённость магнитного поля $($А/м$)$  
$\vec{B}$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$\vec{J}$ — намагниченность $($А/м$)$  

**Особенности** $:$
1. Магнитная восприимчивость и проницаемость зависят от материала.
2. Вектор $\vec{H}$ описывает магнитное поле в вакууме и в веществе.


**<div align="center">Теорема о циркуляции вектора H</div>**

**Формулировка** $:$ циркуляция вектора напряжённости магнитного поля вдоль замкнутого контура равна алгебраической сумме сторонних токов, охватываемых этим контуром:

$$
\oint \vec{H} \cdot d\vec{l} = I
$$

**Где** $:$

$\vec{H}$ — напряжённость магнитного поля $($А/м$)$  
$d\vec{l}$ — элемент длины контура $($м$)$  
$I$ — ток внутри контура $($А$)$  

**Особенности** $:$
1. Теорема используется для расчёта напряжённости магнитного поля.
2. Направление обхода контура определяется правилом правого винта.


**<div align="center">Парамагнетики, диамагнетики, ферромагнетики</div>**

**Парамагнетики** $:$ вещества, у которых магнитная восприимчивость положительна и мала ($\mu > 0$). Примеры: алюминий, платина.

**Диамагнетики** $:$ вещества, у которых магнитная восприимчивость отрицательна ($\mu < 0$). Примеры: медь, серебро.

**Ферромагнетики** $:$ вещества, у которых магнитная восприимчивость положительна и велика ($\mu \gg 1$). Примеры: железо, никель.

**Особенности** $:$
1. Парамагнетики усиливают внешнее магнитное поле.
2. Диамагнетики ослабляют внешнее магнитное поле.
3. Ферромагнетики обладают остаточной намагниченностью.


**<div align="center">Пример: Магнитное поле внутри и вне бесконечно длинного прямого цилиндрического провода с током</div>**

**Условие задачи** $:$ Бесконечно длинный цилиндрический провод радиуса $R$ имеет равномерно распределённый ток $I$. Требуется найти магнитное поле внутри и вне провода.

**Решение** $:$

1. **Внутри провода** $:$

$$
B = \frac{\mu_0 I r}{2\pi R^2}, \quad r \leq R
$$

2. **Вне провода** $:$

$$
B = \frac{\mu_0 I}{2\pi r}, \quad r > R
$$

**Где** $:$

$B$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  
$I$ — сила тока $($А$)$  
$r$ — расстояние от оси провода $($м$)$  
$R$ — радиус провода $($м$)$  

**Особенности** $:$
1. Внутри провода магнитное поле линейно возрастает с расстоянием от оси.
2. Вне провода магнитное поле убывает как $1/r$.
