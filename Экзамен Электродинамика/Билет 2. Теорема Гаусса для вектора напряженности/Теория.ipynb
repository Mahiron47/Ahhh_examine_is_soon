**<div align="center">Поток вектора через поверхность</div>**

**Определение** $:$ поток вектора через поверхность — это скалярная величина, равная интегралу от скалярного произведения вектора напряжённости поля $\vec{E}$ на элементарную площадку $d\vec{S}$ поверхности.

$$
\Phi = \int \vec{E} \cdot d\vec{S}
$$

**Где** $:$

$\Phi$ — поток вектора через поверхность $($Н·м²/Кл$)$  
$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$d\vec{S}$ — элементарная площадка поверхности $($м²$)$  

**Особенности** $:$

1. Поток положителен, если вектор $\vec{E}$ направлен наружу от поверхности.
2. Поток отрицателен, если вектор $\vec{E}$ направлен внутрь поверхности.
3. Поток через замкнутую поверхность зависит от суммы зарядов внутри неё.


**<div align="center">Формулировка и доказательство теоремы</div>**

**Формулировка** $:$ поток вектора напряжённости электрического поля через замкнутую поверхность равен отношению алгебраической суммы зарядов внутри поверхности к электрической постоянной $\varepsilon_0$.

$$
\Phi = \frac{Q}{\varepsilon_0}
$$

**Доказательство** $:$

1. Рассмотрим точечный заряд $q$, создающий поле. Поток через сферическую поверхность радиуса $r$ равен $:$

$$
\Phi = \oint\limits_S \vec{E} \cdot d\vec{S} = \oint\limits_S E \cdot dS \cdot \cos 0.
$$

2. Напряжённость поля $E = k \cdot \frac{q}{r^2}$, элементарная площадка $dS = r^2 \cdot d\Omega$, где $d\Omega$ — телесный угол.

3. Поток через сферу $:$

$$
\Phi = \oint\limits_S k \cdot \frac{q}{r^2} \cdot r^2 \cdot d\Omega = k \cdot q \cdot \oint\limits_S d\Omega = k \cdot q \cdot 4\pi.
$$

4. Подставляем $k = \frac{1}{4\pi\varepsilon_0}$ $:$

$$
\Phi = \frac{q}{\varepsilon_0}.
$$

Для произвольной замкнутой поверхности сумма зарядов внутри неё также даёт поток $\Phi = \frac{Q}{\varepsilon_0}$.


**<div align="center">Теорема Гаусса в дифференциальной форме</div>**

**Формулировка** $:$ дивергенция вектора напряжённости электрического поля равна отношению объёмной плотности заряда к электрической постоянной $\varepsilon_0$.

$$
\nabla \cdot \vec{E} = \frac{\rho}{\varepsilon_0}
$$

**Доказательство** $:$

1. Поток через замкнутую поверхность $:$

$$
\Phi = \oint\limits_S \vec{E} \cdot d\vec{S} = \int\limits_V \nabla \cdot \vec{E} \cdot dV.
$$

2. Согласно теореме Гаусса $:$

$$
\Phi = \frac{Q}{\varepsilon_0} = \frac{1}{\varepsilon_0} \int\limits_V \rho \cdot dV.
$$

3. Сравнивая выражения, получаем $:$

$$
\nabla \cdot \vec{E} = \frac{\rho}{\varepsilon_0}.
$$


**<div align="center">Использование теоремы Гаусса для расчёта полей</div>**

**Поле сферы** $:$

1. Рассмотрим сферу радиуса $R$ с зарядом $Q$. Поток через сферу $:$

$$
\Phi = \oint\limits_S \vec{E} \cdot d\vec{S} = E \cdot 4\pi R^2.
$$

2. Согласно теореме Гаусса $:$

$$
\Phi = \frac{Q}{\varepsilon_0} \Rightarrow E = \frac{Q}{4\pi\varepsilon_0 R^2}.
$$

**Поле шара** $:$

$$
\begin{matrix}
E = \frac{\rho r}{3\varepsilon_0}\qquad r<R\\
\\
E = \frac{Q}{4\pi\varepsilon_0 R^2}\qquad r\ge R
\end{matrix}
$$
Если объёмная плотность заряда шара $\rho$  
**Поле нити** $:$

$$
E = \frac{\lambda}{2\pi\varepsilon_0 r}
$$
Если нить бесконечна с линейной плотностью заряда $\lambda$  
**Поле цилиндра** $:$

$$
E = \frac{\lambda}{2\pi\varepsilon_0 r}
$$
Если линейная плотность заряда цилиндра $\lambda$  
**Поле плоскости** $:$

$$
E = \frac{\sigma}{2\varepsilon_0}.
$$
Если плоскость бесконечна с поверхностной плотностью заряда $\sigma$  
**Поле слоя** $:$

$$
E = \frac{\sigma}{\varepsilon_0}
$$
Если слой бесконечен с поверхностной плотностью заряда $\sigma$
