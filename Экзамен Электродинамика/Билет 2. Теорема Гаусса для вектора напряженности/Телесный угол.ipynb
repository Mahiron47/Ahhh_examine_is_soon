**<div align="center">Определение телесного угла</div>**

Телесный угол — это трёхмерный аналог обычного угла, который используется для измерения площади, занимаемой объектом на сфере относительно её центра. Если обычный угол измеряет "длину дуги" на окружности, то телесный угол измеряет "площадь" на поверхности сферы.

Обычный угол **$\theta$** определяется как отношение длины дуги окружности **$l$** к радиусу этой окружности **$r$**:

$$
\theta = \frac{l}{r}
$$

**Где** $:$

$l$ — длина дуги на окружности.  
$r$ — радиус окружности.  

Телесный угол **$\Omega$** определяется как отношение площади участка сферы **$S$** к квадрату радиуса этой сферы **$r^2$**:

$$
\Omega = \frac{S}{r^2}
$$

**Где** $:$
  
$S$ — элементарная площадь на поверхности сферы.  
$r$ — радиус сферы.
