**<div align="center">Основные экспериментальные факты</div>**

**Факты** $:$
1. Изменение магнитного потока через контур вызывает появление электрического тока в этом контуре.
2. Направление индуцированного тока зависит от направления изменения магнитного потока.
3. Индуцированный ток создаёт магнитное поле, которое противодействует изменению внешнего магнитного поля.

**Примеры** $:$
1. Движение проводника в магнитном поле вызывает появление ЭДС.
2. Изменение силы тока в катушке вызывает ток в соседней катушке.


**<div align="center">Математическая формулировка закона электромагнитной индукции</div>**

**Формулировка** $:$ ЭДС, возникающая в контуре, пропорциональна скорости изменения магнитного потока через этот контур:

$$
\mathcal{E} = -\frac{d\Phi}{dt}
$$

**Где** $:$

$\mathcal{E}$ — электродвижущая сила $($В$)$  
$\Phi$ — магнитный поток $($Вб$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Знак минус отражает правило Ленца.
2. Магнитный поток определяется как $\Phi = \int \vec{B} \cdot d\vec{S}$.


**<div align="center">Правило Ленца</div>**

**Формулировка** $:$ направление индуцированного тока таково, что создаваемое им магнитное поле противодействует изменению внешнего магнитного потока.

**Особенности** $:$
1. Правило Ленца выражает закон сохранения энергии.
2. Индуцированный ток всегда стремится компенсировать изменения внешнего магнитного поля.


**<div align="center">Природа электромагнитной индукции</div>**

**Определение** $:$ электромагнитная индукция — это явление возникновения ЭДС в проводящем контуре при изменении магнитного потока через него.

**Причины** $:$
1. Движение проводника в магнитном поле.
2. Изменение магнитного поля во времени.
3. Изменение площади поверхности.

**Особенности** $:$
1. Электромагнитная индукция лежит в основе работы генераторов и трансформаторов.
2. Индуцированная ЭДС зависит от скорости изменения магнитного потока.


**<div align="center">Закон электромагнитной индукции в интегральной и дифференциальной форме</div>**

**Интегральная форма** $:$

$$
\oint\limits_L \vec{E} \cdot d\vec{l} = -\frac{d\Phi}{dt}
$$

**Дифференциальная форма** $:$

$$
\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}
$$

**Где** $:$

$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$d\vec{l}$ — элемент длины контура $($м$)$  
$\Phi$ — магнитный поток $($Вб$)$  
$\vec{B}$ — магнитная индукция $($Тл$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Интегральная форма используется для расчёта ЭДС в контуре.
2. Дифференциальная форма описывает локальные свойства поля.
