**<div align="center">Плотность тока смещения</div>**

**Определение** $:$ плотность тока смещения $\vec{j}_d$ — это величина, характеризующая изменение электрического поля во времени:

$$
\vec{j}_d = \varepsilon_0 \frac{\partial \vec{E}}{\partial t}
$$

**Где** $:$

$\vec{j}_d$ — плотность тока смещения $($А/м²$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Ток смещения возникает в переменных электрических полях.
2. Он играет ключевую роль в уравнениях Максвелла.


**<div align="center">Уравнения Максвелла в интегральной и дифференциальной форме</div>**

**Интегральная форма** $:$
1. Закон Гаусса для электрического поля:

$$
\oint\limits_S \vec{E} \cdot d\vec{S} = \frac{q}{\varepsilon_0}
$$

2. Закон Гаусса для магнитного поля:

$$
\oint\limits_S \vec{B} \cdot d\vec{S} = 0
$$

3. Закон Фарадея:

$$
\oint\limits_L \vec{E} \cdot d\vec{l} = -\frac{d\Phi_B}{dt}
$$

4. Закон Ампера–Максвелла:

$$
\oint\limits_L \vec{B} \cdot d\vec{l} = \mu_0 \left(I + \varepsilon_0 \frac{d\Phi_E}{dt}\right)
$$

**Дифференциальная форма** $:$
1. Закон Гаусса для электрического поля:

$$
\nabla \cdot \vec{E} = \frac{\rho}{\varepsilon_0}
$$

2. Закон Гаусса для магнитного поля:

$$
\nabla \cdot \vec{B} = 0
$$

3. Закон Фарадея:

$$
\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}
$$

4. Закон Ампера–Максвелла:

$$
\nabla \times \vec{B} = \mu_0 \vec{j} + \mu_0 \varepsilon_0 \frac{\partial \vec{E}}{\partial t}
$$


**<div align="center">Какие закономерности электродинамики выражают уравнения Максвелла?</div>**

**Закономерности** $:$
1. Закон сохранения заряда.
2. Связь электрического и магнитного полей.
3. Возникновение вихревого электрического поля при изменении магнитного поля.
4. Возникновение вихревого магнитного поля при изменении электрического поля или наличии токов.

**Особенности** $:$
1. Уравнения Максвелла описывают электромагнитные волны.
2. Они являются основой классической электродинамики.


**<div align="center">Свойства уравнений Максвелла</div>**

**Свойства** $:$
1. Линейность: уравнения Максвелла линейны относительно полей $\vec{E}$ и $\vec{B}$.
2. Инвариантность: уравнения Максвелла инвариантны относительно преобразований Лоренца.
3. Универсальность: они применимы ко всем электромагнитным явлениям.
4. Согласованность: уравнения Максвелла согласуются с законами сохранения заряда и энергии.

**Особенности** $:$
1. Уравнения Максвелла описывают распространение электромагнитных волн в вакууме и средах.
2. Они связывают электрические и магнитные поля в единую теорию.
