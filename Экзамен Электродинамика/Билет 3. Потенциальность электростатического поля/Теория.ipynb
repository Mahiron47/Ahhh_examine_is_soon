**<div align="center">Работа при перемещении точечного заряда в однородном электрическом поле</div>**

**Определение** $:$ работа при перемещении точечного заряда $q$ в однородном электрическом поле $\vec{E}$ равна произведению заряда, напряжённости поля и проекции перемещения на направление поля.

$$
A = qE\Delta l\cos\alpha=\int\limits_{l_0}^{l_к}\vec E\cdot d\vec l
$$

**Где** $:$  

$A$ — работа $($Дж$)$  
$q$ — заряд $($Кл$)$  
$E$ — напряжённость электрического поля $($Н/Кл$)$  
$\Delta l$ — перемещение $($м$)$  
$\alpha$ — угол между $\vec{E}$ и направлением перемещения  


**<div align="center">Работа при перемещении точечного заряда в поле другого неподвижного точечного заряда</div>**

**Определение** $:$ работа при перемещении точечного заряда $q$ в электрическом поле равна изменению потенциальной энергии системы.

$$
A = q\left(\phi_0 - \phi_к\right) = q\left(\frac{Q}{4\pi\varepsilon_0 l_0} - \frac{Q}{4\pi\varepsilon_0 l_к}\right)
$$

**Где** $:$  

$q$ — заряд $($Кл$)$  
$Q$ — точечный заряд $($Кл$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$r_1$, $r_2$ — расстояния от заряда $Q$ до начальной и конечной точек перемещения $($м$)$  
$A$ — работа $($Дж$)$  
$\phi_к$, $\phi_0$ — расстояния от заряда $Q$ до начальной и конечной точек перемещения $($Дж$)$

**<div align="center">Циркуляция электростатического поля по произвольному замкнутому контуру</div>**

**Определение** : циркуляция электростатического поля $\vec{E}$ по замкнутому контуру равна нулю.

$$
\oint\limits_L \vec{E} \cdot d\vec{l} = 0
$$
**Где** $:$

$\vec{E}$ — вектор напряжённости электростатического поля (Н/Кл)  
$d\vec{l}$ — элемент длины контура, направленный по касательной к контуру (м)  
$L$ — замкнутый контур, вдоль которого вычисляется циркуляция (м)  

**Особенности** :  
Электростатическое поле является потенциальным, поэтому его циркуляция по замкнутому контуру всегда равна нулю.

**<div align="center">Разность потенциалов. Потенциал. Потенциал поля точечного заряда. Принцип суперпозиции</div>**
**<div align="center">Потенциал точечного заряда</div>**

**Разность потенциалов** : разность потенциалов между двумя точками равна работе, совершаемой внешними силами при перемещении единичного положительного заряда из одной точки в другую.

$$
\varphi_0 - \varphi_к = \frac{A}{q}
$$

**Где** $:$ 

$\varphi_0$ и $\varphi_к$ — это начальный и конечный потенциал (Дж)  
$A$ — обозначает работу, выполненную над системой или внутри системы (Дж)  
$q$ — заряд (Кл)  

**Потенциал** : скалярная величина, равная сумме начального потенциала и работы по перемещению из конечной точки в изначальную делённую на заряд

$$
\varphi(\vec r)=\varphi_0+\int\limits_{\vec r}^{\vec{r_0}}\vec E\cdot d\vec l
$$

**Потенциал поля точечного заряда** : скалярная величина, равная работе, совершаемой внешними силами при перемещении единичного положительного заряда из данной точки в бесконечность.

$$
\varphi = \frac{Q}{4\pi\varepsilon_0r}
$$

**Где** $:$  

$\varphi$ — потенциал (В)  
$Q$ — заряд (Кл)  
$r$ — расстояние до заряда $Q$  

**Принцип суперпозиции** $:$ потенциал системы зарядов равен сумме потенциалов, создаваемых каждым зарядом.

$$
\varphi = \frac{1}{4\pi\varepsilon_0}\sum\limits_{i=0}\frac{Q_i}{r_i}
$$

**<div align="center">Связь потенциала с напряжённостью электростатического поля</div>**

**Определение** : напряжённость электростатического поля равна градиенту потенциала с обратным знаком.

$$
\vec{E} = - \nabla \varphi
$$

**Пример** : расчёт потенциала и напряжённости поля на оси однородно заряженного кольца.  
Для кольца радиуса $R$ с зарядом $Q$ потенциал в точке на оси кольца на расстоянии $z$ от центра :

$$
\varphi = k \cdot \frac{Q}{\sqrt{R^2 + z^2}}
$$

Напряжённость поля :

$$
E_z = - \frac{\partial \varphi}{\partial z} = k \cdot \frac{Q \cdot z}{(R^2 + z^2)^{3/2}}
$$
