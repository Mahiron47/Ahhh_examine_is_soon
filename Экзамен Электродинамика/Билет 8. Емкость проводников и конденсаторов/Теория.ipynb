**<div align="center">Емкость уединенного проводника. Емкость уединенного проводящего шара</div>**

**Определение** $:$ емкость уединенного проводника — это скалярная величина, характеризующая способность проводника накапливать электрический заряд при заданном потенциале.

$$
C = \frac{q}{\varphi}
$$

**Где** $:$

$C$ — емкость проводника $($Ф$)$  
$q$ — заряд проводника $($Кл$)$  
$\varphi$ — потенциал проводника $($В$)$  

**Емкость уединенного проводящего шара** $:$

$$
C = 4\pi\varepsilon_0 R
$$

**Где** $:$

$R$ — радиус шара $($м$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  

**Особенности** $:$
1. Емкость проводника зависит только от его геометрии.
2. Для шара емкость пропорциональна его радиусу.


**<div align="center">Конденсатор. Емкость конденсатора</div>**

**Определение** $:$ конденсатор — это два проводника произвольной формы разделённые диэлектриком

**Емкость конденсатора** $:$ емкость конденсатора определяется как отношение заряда на одном из проводников к разности потенциалов между ними:

$$
C = \frac{Q}{\varphi_+-\varphi_-}
$$

**Где** $:$

$C$ — емкость конденсатора $($Ф$)$  
$Q$ — заряд на проводнике $($Кл$)$  
$\varphi_+-\varphi_-$ — разность потенциалов между проводниками $($В$)$  

**Особенности** $:$
1. Емкость зависит от геометрии проводников и свойств диэлектрика.
2. Конденсаторы используются для накопления энергии и сглаживания колебаний напряжения.


**<div align="center">Емкость плоского конденсатора</div>**

**Определение** $:$ плоский конденсатор состоит из двух параллельных проводящих пластин, разделенных диэлектриком.

**Емкость** $:$

$$
C = \frac{\varepsilon_0 \varepsilon S}{d}
$$

**Где** $:$

$C$ — емкость конденсатора $($Ф$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\varepsilon$ — диэлектрическая проницаемость материала $($безразмерная$)$  
$S$ — площадь пластин $($м²$)$  
$d$ — расстояние между пластинами $($м$)$  

**Особенности** $:$
1. Емкость увеличивается с увеличением площади пластин и диэлектрической проницаемости.
2. Емкость уменьшается с увеличением расстояния между пластинами.


**<div align="center">Емкость сферического конденсатора</div>**

**Определение** $:$ сферический конденсатор состоит из двух концентрических сферических проводников, разделенных диэлектриком.

**Емкость** $:$

$$
C = \frac{4\pi\varepsilon_0 \varepsilon R_1 R_2}{R_2 - R_1}
$$

**Где** $:$

$C$ — емкость конденсатора $($Ф$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  
$\varepsilon$ — диэлектрическая проницаемость материала $($безразмерная$)$  
$R_1$ — радиус внутренней сферы $($м$)$  
$R_2$ — радиус внешней сферы $($м$)$  

**Особенности** $:$
1. Емкость зависит от радиусов сфер и диэлектрической проницаемости.
2. Емкость увеличивается с уменьшением расстояния между сферами.
