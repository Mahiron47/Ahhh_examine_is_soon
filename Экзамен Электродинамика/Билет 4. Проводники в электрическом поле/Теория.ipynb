**<div align="center">Явление электростатической индукции</div>**

**Определение** $:$ явление перераспределения зарядов в проводнике под действием внешнего электростатического поля.  

**Особенности** $:$  
1. Внешнее поле вызывает движение свободных зарядов внутри проводника.  
2. На поверхности проводника образуются индуцированные заряды, компенсирующие действие внешнего поля внутри проводника.  
3. В результате внутри проводника устанавливается равновесие, при котором напряжённость электростатического поля равна нулю.  

**<div align="center">Напряжённость электростатического поля внутри проводника</div>**

**Определение** $:$ напряжённость электростатического поля внутри проводника в состоянии электростатического равновесия равна нулю.  

**Особенности** $:$  
1. Внутри проводника свободные заряды перераспределяются так, чтобы компенсировать действие внешнего поля.  
2. Если напряжённость внутри проводника не равна нулю, заряды продолжают двигаться, пока не установится равновесие.  

**<div align="center">Потенциал внутри проводника</div>**

**Определение** $:$ потенциал внутри проводника в состоянии электростатического равновесия одинаков во всех точках.  

**Особенности** $:$  
1. Потенциал проводника является скалярной величиной и не зависит от положения внутри проводника.  
2. На поверхности проводника потенциал также одинаков и равен потенциалу внутри проводника. 

**<div align="center">Напряжённость и потенциал в полости внутри проводника</div>**

**Определение** $:$ если внутри проводника имеется полость, то:  
1. **Напряжённость** электростатического поля в полости равна нулю, если в ней отсутствуют заряды.  
2. **Потенциал** в полости одинаков во всех точках и равен потенциалу проводника.  

**Особенности** $:$  
1. Если в полости находятся заряды, они создают поле внутри полости, но не влияют на распределение зарядов на поверхности проводника.  
2. Внешнее поле не проникает в полость проводника.  

**<div align="center">Заряд внутри и на поверхности проводника</div>**

**Определение** $:$ в состоянии электростатического равновесия:  
1. Заряд внутри проводника отсутствует.  
2. Весь заряд проводника сосредоточен на его поверхности.  

**Особенности** $:$  
1. Заряды на поверхности распределяются так, чтобы компенсировать действие внешнего поля внутри проводника.  
2. Распределение зарядов на поверхности зависит от формы проводника и внешнего поля.  

**<div align="center">Напряжённость электростатического поля вблизи поверхности проводника</div>**

**Определение** $:$ напряжённость электростатического поля вблизи поверхности проводника направлена перпендикулярно к поверхности.  

**Формула** $:$  
$$
E = \frac{\sigma}{\varepsilon_0}
$$  

**Где** $:$  

$E$ — напряжённость электростатического поля $($Н/Кл$)$  
$\sigma$ — поверхностная плотность заряда $($Кл/м²$)$  
$\varepsilon_0$ — электрическая постоянная $($Ф/м$)$  

**Особенности** $:$  
1. Напряжённость поля пропорциональна поверхностной плотности заряда.  
2. Напряжённость поля максимальна на участках поверхности с наибольшей кривизной.  
3. Напряжённость поля вне проводника зависит от внешнего поля и распределения зарядов на поверхности.
