**<div align="center">Квазистационарные токи</div>**

**Определение** $:$ квазистационарные токи — это токи, при которых распределение заряда и магнитного поля изменяется достаточно медленно, чтобы можно было пренебречь запаздыванием электромагнитных взаимодействий.

**Особенности** $:$
1. Квазистационарные токи описываются законами постоянного тока.
2. Применимы для низкочастотных процессов.
3. Электромагнитные волны не учитываются.


**<div align="center">Резистор в цепи переменного тока</div>**

**Особенности** $:$
1. Напряжение и ток в резисторе находятся в фазе.
$$
i=I_m\omega\sin{(\omega t+\alpha)}\qquad u=U_m\cos{(\omega t+\alpha)}
$$
2. Закон Ома для переменного тока:

$$
I = \frac{U}{R}
$$

**Где** $:$

$I$ — сила тока $($А$)$  
$U$ — напряжение $($В$)$  
$R$ — сопротивление $($Ом$)$  

**Особенности** $:$
1. Мощность, выделяемая на резисторе:

$$
P = I^2 R
$$


**<div align="center">Конденсатор в цепи переменного тока</div>**

**Особенности** $:$
1. Ток опережает напряжение на $90^\circ$.
2. Ёмкостное сопротивление:

$$
X_C = \frac{1}{\omega C}
$$

**Где** $:$

$X_C$ — ёмкостное сопротивление $($Ом$)$  
$\omega$ — циклическая частота $($рад/с$)$  
$C$ — ёмкость $($Ф$)$  

**Закон Ома для конденсатора** $:$

$$
I = \frac{U}{X_C}
$$


**<div align="center">Индуктивность в цепи переменного тока</div>**

**Особенности** $:$
1. Напряжение опережает ток на $90^\circ$.
2. Индуктивное сопротивление:

$$
X_L = \omega L
$$

**Где** $:$

$X_L$ — индуктивное сопротивление $($Ом$)$  
$\omega$ — циклическая частота $($рад/с$)$  
$L$ — индуктивность $($Гн$)$  

**Закон Ома для индуктивности** $:$

$$
I = \frac{U}{X_L}
$$


**<div align="center">Метод векторных диаграмм. Резонанс в колебательном контуре</div>**

**Метод векторных диаграмм** $:$
1. Используется для анализа цепей переменного тока.
2. Ток и напряжение представляются в виде векторов на комплексной плоскости.
3. Суммарное напряжение определяется геометрически.

**Резонанс** $:$
1. Резонанс возникает, когда ёмкостное и индуктивное сопротивления равны:

$$
X_C = X_L \quad \Rightarrow \quad \omega = \frac{1}{\sqrt{LC}}
$$

2. При резонансе ток достигает максимального значения.


**<div align="center">Метод комплексных амплитуд</div>**

**Определение** $:$ метод комплексных амплитуд используется для анализа цепей переменного тока, представляя ток и напряжение в виде комплексных чисел.

**Формулы** $:$
1. Комплексное напряжение:

$$
\tilde{U} = U_0 e^{j\omega t}
$$

2. Комплексный ток:

$$
\tilde{I} = I_0 e^{j\omega t}
$$

**Особенности** $:$
1. Метод упрощает расчёты в цепях переменного тока.
2. Фазовые соотношения между током и напряжением легко определяются.
