**<div align="center">Дифференциальное уравнение колебательного контура</div>**

**Определение** $:$ колебательный контур состоит из конденсатора и катушки индуктивности, соединённых последовательно. Дифференциальное уравнение для заряда $q(t)$ на конденсаторе имеет вид:

$$
L \frac{d^2q}{dt^2} + R \frac{dq}{dt} + \frac{q}{C} = 0
$$

**Где** $:$

$L$ — индуктивность катушки $($Гн$)$  
$R$ — сопротивление контура $($Ом$)$  
$C$ — ёмкость конденсатора $($Ф$)$  
$q(t)$ — заряд на конденсаторе $($Кл$)$  
$t$ — время $($с$)$  

**Особенности** $:$
1. Уравнение описывает взаимодействие электрического и магнитного полей в контуре.
2. Решение уравнения зависит от величины сопротивления $R$.


**<div align="center">Свободные колебания в колебательном контуре</div>**

**Определение** $:$ свободные колебания возникают в идеальном контуре ($R = 0$), когда энергия передаётся между электрическим полем конденсатора и магнитным полем катушки.

**Решение уравнения** $:$

$$
q(t) = q_0 \cos(\omega t)
$$

**Где** $:$

$q_0$ — начальный заряд на конденсаторе $($Кл$)$  
$\omega$ — собственная циклическая частота контура $($рад/с$)$  

**Собственная частота** $:$

$$
\omega = \frac{1}{\sqrt{LC}}
$$

**Особенности** $:$
1. Колебания являются гармоническими.
2. Энергия сохраняется и передаётся между компонентами контура.


**<div align="center">Затухающие колебания в колебательном контуре</div>**

**Определение** $:$ затухающие колебания возникают в реальном контуре ($R > 0$), когда часть энергии рассеивается в виде тепла.

**Решение уравнения** $:$

$$
q(t) = q_0 e^{-\frac{R}{2L}t} \cos(\omega' t)
$$

**Где** $:$

$q_0$ — начальный заряд на конденсаторе $($Кл$)$  
$\omega'$ — циклическая частота затухающих колебаний $($рад/с$)$  

**Циклическая частота** $:$

$$
\omega' = \sqrt{\frac{1}{LC} - \frac{R^2}{4L^2}}
$$

**Особенности** $:$
1. Амплитуда колебаний экспоненциально уменьшается со временем.
2. Энергия рассеивается в сопротивлении контура.
