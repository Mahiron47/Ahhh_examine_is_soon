**<div align="center">Индуктивность проводов</div>**

**Определение** $:$ индуктивность — это скалярная величина, характеризующая способность проводника или контура с током создавать магнитный поток, пропорциональный силе тока.

$$
L = \frac{\Phi}{I}
$$

**Где** $:$

$L$ — индуктивность $($Гн$)$  
$\Phi$ — магнитный поток $($Вб$)$  
$I$ — сила тока $($А$)$  

**Особенности** $:$
1. Индуктивность зависит от геометрии проводника и свойств среды.
2. Для длинного соленоида индуктивность определяется как:

$$
L = \mu_0 \mu n^2 V
$$

**Где** $:$

$n$ — число витков на единицу длины $($м⁻¹$)$  
$V$ — объём соленоида $($м³$)$  
$\mu$ — магнитная проницаемость среды.


**<div align="center">Явления при замыкании и размыкании тока</div>**

**Замыкание тока** $:$ при замыкании цепи ток возрастает от нуля до установившегося значения. В этот момент возникает ЭДС самоиндукции, препятствующая изменению тока:

$$
\mathcal{E}_{\text{сам}} = -L \frac{dI}{dt}
$$

**Размыкание тока** $:$ при размыкании цепи ток уменьшается до нуля. ЭДС самоиндукции препятствует уменьшению тока.

**Особенности** $:$
1. ЭДС самоиндукции пропорциональна скорости изменения тока.
2. Энергия магнитного поля проводника преобразуется в тепло.


**<div align="center">Магнитная энергия токов</div>**

**Определение** $:$ магнитная энергия токов — это энергия, накопленная в магнитном поле, создаваемом токами.

$$
W = \frac{1}{2}LI^2
$$

**Где** $:$

$W$ — магнитная энергия $($Дж$)$  
$L$ — индуктивность $($Гн$)$  
$I$ — сила тока $($А$)$  

**Особенности** $:$
1. Магнитная энергия пропорциональна квадрату силы тока.
2. Энергия накапливается в магнитном поле проводника или контура.


**<div align="center">Локализация магнитной энергии в пространстве</div>**

**Определение** $:$ плотность магнитной энергии — это энергия, сосредоточенная в единице объёма пространства, заполненного магнитным полем.

$$
w = \frac{B^2}{2\mu_0}
$$

**Где** $:$

$w$ — плотность магнитной энергии $($Дж/м³$)$  
$B$ — магнитная индукция $($Тл$)$  
$\mu_0$ — магнитная постоянная $($Н/м²$)$  

**Локализация энергии** $:$ энергия магнитного поля локализована в пространстве, где существует поле. Для расчёта полной энергии используется интеграл:

$$
W = \int\limits_V w \cdot dV = \int\limits_V \frac{B^2}{2\mu_0} \cdot dV
$$

**Особенности** $:$
1. Плотность энергии пропорциональна квадрату магнитной индукции.
2. Полная энергия поля определяется интегрированием плотности по объёму.
