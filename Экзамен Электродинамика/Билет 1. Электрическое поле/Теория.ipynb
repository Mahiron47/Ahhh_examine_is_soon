**<div align="center">Электрический заряд</div>**

**Электрический заряд** — скалярная величина, определяющая способность тел участвовать в электромагнитном взаимодействии

**Кулон** — это заряд, который проходит через поперечное сечение проводника за секунду при силе тока в ампер.
$$
1 \text{ Кл} = 1 \text{ А} \cdot 1 \text{ с}
$$

**<div align="center">Свойства электрического заряда</div>**

1.  **Взаимодействие** $:$ существует два вида зарядов — положительные и отрицательные. Одноимённые заряды отталкиваются, разноимённые — притягиваются.
2.  **Закон сохранения** $:$ полный заряд замкнутой системы постоянен.
3.  **Релятивистская инвариантность** $:$ величина заряда не зависит от системы отсчёта и его скорости.
4.  **Дискретность** $:$ заряд любого тела кратен элементарному заряду $e \approx 1.6 \cdot 10^{-19} \text{ Кл}$.
    *   Заряд электрона $:$ $-e$
    *   Заряд протона $:$ $+e$



**Пробный заряд** — это точечный заряд, который вносится в электростатическое поле для его исследования и удовлетворяет двум условиям:

1.  Его величина настолько мала, что он не вызывает заметного перераспределения исследуемых зарядов.
2.  Его размеры пренебрежимо малы по сравнению с расстояниями до зарядов, создающих поле.


**<div align="center">Закон Кулона</div>**

**Формулировка** $:$ сила взаимодействия двух точечных зарядов в вакууме прямо пропорциональна произведению величин этих зарядов и обратно пропорциональна квадрату расстояния между ними.

$$
F = k \cdot \frac{|q_1q_2|}{r^2}
$$

**Где** $:$

$F$ — сила взаимодействия зарядов $($Н$)$  
$q_1, q_2$ — величины зарядов $($Кл$)$  
$r$ — расстояние между зарядами $($м$)$  
$k$ — коэффициент пропорциональности $($Н·м²/Кл²$)$, в вакууме $k = 9 \cdot 10^9$

**Примечание** $:$ коэффициент $k$ можно выразить через электрическую постоянную $\varepsilon_0$ $($Ф/м$)$:
$$
k = \frac{1}{4 \pi \varepsilon_0}
$$
где $\varepsilon_0 \approx 8.85 \cdot 10^{-12}$ $($Ф/м$)$.

**Особенности** $:$

1. Закон применим только для точечных зарядов или сферически симметричных распределений зарядов.
2. Сила взаимодействия направлена вдоль прямой, соединяющей заряды.
3. Сила притяжения или отталкивания зависит от знаков зарядов:
	* Одноимённые заряды отталкиваются.
	* Разноимённые заряды притягиваются.

**<div align="center">Напряжённость электрического поля неподвижного точечного заряда</div>**

**Формулировка** $:$ напряжённость электрического поля в данной точке пространства определяется как сила, действующая на пробный заряд, помещённый в эту точку, делённая на величину этого заряда.

$$
\vec{E} = \frac{\vec{F}}{q}
$$

Для неподвижного точечного заряда напряжённость электрического поля выражается формулой $:$

$$
E = k \cdot \frac{|q|}{r^2}
$$

**Где** $:$

$\vec{E}$ — напряжённость электрического поля $($Н/Кл$)$  
$\vec{F}$ — сила, действующая на пробный заряд $($Н$)$  
$q$ — величина пробного заряда $($Кл$)$  
$E$ — модуль напряжённости электрического поля $($Н/Кл$)$  
$q$ — величина заряда, создающего поле $($Кл$)$  
$r$ — расстояние от заряда до точки, где определяется напряжённость $($м$)$  
$k$ — коэффициент пропорциональности $($Н·м²/Кл²$)$, в вакууме $k = 9 \cdot 10^9$

**Особенности** $:$

1. Напряжённость электрического поля — векторная величина, её направление совпадает с направлением силы, действующей на положительный пробный заряд.
2. Напряжённость уменьшается с увеличением расстояния от заряда.
3. Для отрицательного заряда направление напряжённости направлено к заряду, для положительного — от заряда.

**<div align="center">Принцип суперпозиции электрических полей</div>**

**Формулировка** $:$ напряжённость электрического поля, создаваемого системой зарядов в данной точке пространства, равна векторной сумме напряжённостей полей, создаваемых каждым зарядом в этой точке.

$$
\vec{E} = \sum \vec{E}_i
$$

**Где** $:$

$\vec{E}$ — результирующая напряжённость электрического поля $($Н/Кл$)$  
$\vec{E}_i$ — напряжённость электрического поля, создаваемого $i$-м зарядом $($Н/Кл$)$  

**Особенности** $:$

1. Принцип суперпозиции применим для любых систем зарядов, независимо от их количества и расположения.
2. Напряжённости электрических полей складываются как векторные величины, учитывая их направление.
3. Принцип суперпозиции позволяет рассчитать поле сложных систем зарядов, разбивая их на отдельные элементы.

**<div align="center">Пример: Вектор напряжённости на оси однородно заряженного кольца</div>**

**Условие задачи** $:$ Однородно заряженное кольцо радиуса $R$ имеет заряд $Q$. Требуется найти напряжённость электрического поля $\vec{E}$ на оси кольца на расстоянии $z$ от его центра.

**Решение** $:$

1. Рассмотрим элемент кольца с зарядом $dq$. Напряжённость электрического поля, создаваемого этим элементом, направлена вдоль прямой, соединяющей элемент кольца и точку на оси.  
2. Из-за симметрии кольца горизонтальные компоненты напряжённости электрического поля взаимно уничтожаются, а вертикальные компоненты складываются.  
3. Напряжённость электрического поля в точке на оси кольца определяется формулой $:$

$$
E = k \cdot \frac{Qz}{(R^2 + z^2)^{3/2}}
$$

**Где** $:$  
$E$ — модуль напряжённости электрического поля $($Н/Кл$)$  
$k$ — коэффициент пропорциональности $($Н·м²/Кл²$)$, в вакууме $k = 9 \cdot 10^9$  
$Q$ — полный заряд кольца $($Кл$)$  
$z$ — расстояние от центра кольца до точки на оси $($м$)$  
$R$ — радиус кольца $($м$)$  

**Вывод формулы** $:$

1. Напряжённость от элемента кольца $dq$ на расстоянии $r$ от точки на оси $z$ $:$

$$
dE = k \cdot \frac{dq}{r^2}
$$

где $r = \sqrt{R^2 + z^2}$.

2. Вертикальная компонента напряжённости $dE_z$ $:$

$$
dE_z = dE \cdot \frac{z}{r} = k \cdot \frac{dq \cdot z}{r^3}.
$$

3. Суммируем вклад всех элементов кольца $:$

$$
E_z = \int\limits_0^{E_z} dE_z = k \cdot \frac{z}{r^3} \int\limits_0^{Q} dq = k \cdot \frac{Qz}{(R^2 + z^2)^{3/2}}.
$$

**Особенности** $:$

1. Напряжённость уменьшается с увеличением расстояния $z$ от кольца.  
2. На оси кольца напряжённость направлена вдоль оси.  
3. Если $z \gg R$, формула упрощается до $E \approx k \cdot \frac{Q}{z^2}$, что соответствует напряжённости точечного заряда.  

**<div align="center">Силовые линии электрического поля</div>**

**Определение** $:$ силовые линии электрического поля — это воображаемые линии, касательные к которым в каждой точке совпадают с направлением вектора напряжённости электрического поля $\vec{E}$.

**Особенности** $:$

1. **Направление** $:$ силовые линии направлены от положительных зарядов к отрицательным.  
2. **Плотность линий** $:$ плотность силовых линий (их количество на единицу площади) пропорциональна величине напряжённости электрического поля.  
3. **Замкнутость** $:$ силовые линии начинаются на положительных зарядах и заканчиваются на отрицательных, либо уходят в бесконечность.  
4. **Пересечение** $:$ силовые линии не пересекаются, так как в каждой точке пространства напряжённость электрического поля имеет единственное направление.  

**Примеры** $:$

1. **Точечный заряд** $:$ силовые линии радиально расходятся от положительного заряда и сходятся к отрицательному.  
2. **Диполь** $:$ силовые линии выходят из положительного заряда и входят в отрицательный, образуя симметричную структуру.  
3. **Однородное поле** $:$ силовые линии параллельны и равномерно распределены.

**Применение** $:$

1. Визуализация электрического поля.  
2. Анализ направления и величины напряжённости поля.  
3. Решение задач электростатики.

**Примечание** $:$ силовые линии — это графическое представление, они не существуют физически.
