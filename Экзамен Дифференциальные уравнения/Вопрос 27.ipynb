Система линейных неоднородных дифференциальных уравнений 1-го порядка с постоянными коэффициентами.  Теорема о общем решении системы линейных неоднородных дифференциальных уравнений 1-го порядка с постоянными коэффициентами (с доказательством).
</div>

Рассмотрим систему линейных неоднородных дифференциальных уравнений первого порядка с постоянными коэффициентами в матричной форме $:$

$$
\frac{d\mathbf{X}}{dt} = A\mathbf{X} + \mathbf{F}(t),
$$

где $:$
- $\mathbf{X}(t)$ — вектор-функция неизвестных размерности $n \times 1$,
- $A$ — постоянная матрица коэффициентов размерности $n \times n$,
- $\mathbf{F}(t)$ — вектор-функция неоднородностей размерности $n \times 1$.


**Теорема $:$**  
Общее решение $\mathbf{X}(t)$ такой системы имеет вид $:$

$$
\mathbf{X}(t) = \mathbf{X}_{\text{одн}}(t) + \mathbf{X}_{\text{част}}(t),
$$

где $:$
- $\mathbf{X}_{\text{одн}}(t)$ — общее решение соответствующей однородной системы $\frac{d\mathbf{X}}{dt} = A\mathbf{X}$,
- $\mathbf{X}_{\text{част}}(t)$ — частное решение неоднородной системы.

**Доказательство $:$** решение однородной системы $\frac{d\mathbf{X}}{dt} = A\mathbf{X}$ имеет вид $:$

$$
   \mathbf{X}_{\text{одн}}(t) = e^{At}\mathbf{C},
$$

где $e^{At}$ — матричная экспонента, $\mathbf{C}$ — произвольный постоянный вектор.

Будем искать частное решение неоднородной системы в виде $:$

$$
   \mathbf{X}_{\text{част}}(t) = e^{At}\mathbf{C}(t),
$$

где $\mathbf{C}(t)$ — некоторая вектор-функция, которую нужно определить. Подставим это в исходную систему $:$

$$
   \frac{d}{dt}\left(e^{At}\mathbf{C}(t)\right) = A e^{At}\mathbf{C}(t) + \mathbf{F}(t).
$$

Раскрывая производную, получаем $:$

$$
   A e^{At}\mathbf{C}(t) + e^{At}\frac{d\mathbf{C}(t)}{dt} = A e^{At}\mathbf{C}(t) + \mathbf{F}(t).
$$

Сокращая одинаковые слагаемые, находим $:$

$$
   e^{At}\frac{d\mathbf{C}(t)}{dt} = \mathbf{F}(t).
$$

Умножая слева на $e^{-At}$, получаем $:$

$$
   \frac{d\mathbf{C}(t)}{dt} = e^{-At}\mathbf{F}(t).
$$

Интегрируя, находим $\mathbf{C}(t)$ $:$

$$
   \mathbf{C}(t) = \int e^{-At}\mathbf{F}(t) \, dt + \mathbf{C}_0,
$$

где $\mathbf{C}_0$ — постоянный вектор. Тогда частное решение $:$

$$
   \mathbf{X}_{\text{част}}(t) = e^{At}\int e^{-At}\mathbf{F}(t) \, dt.
$$

(Константу $\mathbf{C}_0$ можно опустить, так как она уже учтена в общем решении однородной системы.)

Объединяя решения, получаем $:$

$$
   \mathbf{X}(t) = e^{At}\mathbf{C} + e^{At}\int e^{-At}\mathbf{F}(t) \, dt.
$$

Это и есть общее решение системы.

**Замечание $:$**  
Если $\mathbf{F}(t)$ имеет специальный вид (например, квазиполином), то частное решение можно найти методом неопределённых коэффициентов, не прибегая к вариации постоянных.

Таким образом, общее решение системы линейных неоднородных дифференциальных уравнений складывается из общего решения однородной системы и частного решения неоднородной.
