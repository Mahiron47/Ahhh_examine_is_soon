Метод Эйлера решения линейных однородные дифференциальные уравнения с постоянными коэффициентами. Характеристическое уравнение. Построение общего решения уравнений в различных случаях корней характеристического уравнения (с обоснованием).
</div>

Уравнение Эйлера имеет вид $:$
$$
y^{(n)} + a_{n-1}y^{(n-1)} + \dots + a_1y' + a_0y = 0,
$$
где $a_i$ - постоянные коэффициенты, $n$ - порядок уравнения.

**Характеристическое уравнение**

Решение ищется в виде экспоненты $y = e^{kx}$. Подставляя в уравнение, получаем $:$

$$
k^n e^{kx} + a_{n-1} k^{n-1} e^{kx} + \dots + a_1 k e^{kx} + a_0 e^{kx} = 0.
$$

Так как $e^{kx} \neq 0$, можно сократить на $e^{kx}$ $:$

$$
k^n + a_{n-1} k^{n-1} + \dots + a_1 k + a_0 = 0.
$$

Это уравнение называется **характеристическим уравнением**. Его корни определяют вид общего решения дифференциального уравнения.

**Случай 1: Все корни действительные и различные**

Пусть $k_1, k_2, \dots, k_n$ — различные действительные корни. Тогда общее решение имеет вид $:$

$$
y(x) = C_1 e^{k_1 x} + C_2 e^{k_2 x} + \dots + C_n e^{k_n x},
$$

где $C_1, C_2, \dots, C_n$ — произвольные постоянные.

**Обоснование:**  
Каждая экспонента $e^{k_i x}$ является решением, а линейная комбинация решений также является решением (в силу линейности и однородности уравнения).

**Случай 2: Корни действительные, но есть кратные**

Пусть $k$ — корень кратности $m$. Тогда общее решение имеет вид $:$

$$
y(x) = C_1 e^{k x} + C_2 xe^{k x} + \dots + C_n x^{m-1}e^{k x},
$$

Если есть несколько кратных корней, для каждого из них строится аналогичная система решений.

**Обоснование:**  
Можно показать, что функции $x^j e^{k x}$ ($j = 0, 1, \dots, m-1$) являются линейно независимыми решениями.

**Случай 3: Есть комплексные корни**

Комплексные корни характеристического уравнения встречаются комплексно-сопряжёнными парами $\alpha \pm i \beta$. Каждой такой паре соответствует два действительных решения:

$$
e^{\alpha x} \cos(\beta x) \quad \text{и} \quad e^{\alpha x} \sin(\beta x).
$$

Если комплексный корень имеет кратность $m$, то добавляются множители $x, x^2, \dots, x^{m-1}$.

**Обоснование:**  
Используется формула Эйлера $e^{i \beta x} = \cos(\beta x) + i \sin(\beta x)$, позволяющая перейти от комплексных решений к действительным.
