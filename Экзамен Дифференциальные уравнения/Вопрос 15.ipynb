Линейное дифференциальное уравнение $n$-ого порядка
$$
y^{(n)}+P_1(x)y^{(n-1)}+P_2(x)y^{(n-2)}+\,...\,+P_n(x)y=Q(x)\\
(i\in[\,0,\,1,\,2,\,...,\,n\,]: P_i(x),\,Q(x))\ - \ \text{непрерывные}
$$

Выражаем всё относительно старшей производной
$$
y^{(n)}=-P_1(x)y^{(n-1)}-P_2(x)y^{(n-2)}-\,...\,-P_n(x)y+Q(x)=\\=f(x,\,y)
$$
Легко можно показать, что **Т о существовании и единственности** для $n$-ого порядка выполняется из начального условия непрерывности и $\displaystyle\left|\frac{\delta f}{\delta y^{(i)}}\right|\le-P_i(x)$ .
Введём понятие линейного оператора
$$
L=\frac{d^n}{dx^n}+P_1(x)\frac{d^{n-1}}{dx^{n-1}}+\,...\,+P_n(x)y\ - \ \text{линейный}\ \text{оператор}
$$
Перепишем уравнение в форме линейного оператора
$$
Ly=Q(x)
$$
Линейные однородные дифф уравнения $n$-ого порядка
$$
Ly=0
$$
Линейный оператор обладает свойствами
1. $LCy=CLy$
2. $L(y_1+y_2)=Ly_1+Ly_2$

**Т (О решении однородного линейного дифф уравнения первым образом)** $:$ если $y(x)\ -$ решение однородного линейного дифф уравнения $Ly=0$ $,$ то и $Cy(x)\ -$ решение $,$ если $(y_1(x),\,y_2(x),\,...,\,y_n(x))\ -$ решения однородного линейного дифф уравнения $Ly=0$ $,$ то и $\displaystyle\sum^k_{i=1}{C_iy_i}\ -$ решения $Ly=0$  

Док-во $:$ воздействуем линейным оператором на
$$
LCy=CLy=C\times0=0
$$
Из чего следует первая часть **Т** $,$ вторая же является комбинацией линейных свойств линейного оператора
$$
L\sum^k_{i=1}{C_iy_i}=\sum^k_{i=1}C_iLy_i=0\times\sum^k_{i=1}C_i=0
$$
что и требовалось доказать. $\blacksquare$
