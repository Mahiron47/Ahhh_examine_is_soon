**Векторная функция** — это функция вида
$$
\mathbf{x}(t) = \begin{pmatrix} x_1(t) \\ x_2(t) \\ \vdots \\ x_n(t) \end{pmatrix}
$$
где $x_i(t)$ — скалярные функции.


**Линейная независимость векторных функций** $:$ система векторных функций $\mathbf{x}_1(t), \mathbf{x}_2(t), \dots, \mathbf{x}_n(t)$ называется **линейно независимой** на интервале $I$, если равенство
$$
C_1 \mathbf{x}_1(t) + C_2 \mathbf{x}_2(t) + \dots + C_n \mathbf{x}_n(t) = \mathbf{0} \quad \forall t \in I
$$
возможно только при $C_1 = C_2 = \dots = C_n = 0$.

**Определитель Вронского для векторных функций**

Пусть даны $n$ векторных функций $\mathbf{x}_1(t), \mathbf{x}_2(t), \dots, \mathbf{x}_n(t)$, где каждый вектор имеет $n$ компонент. **Определитель Вронского** (вронскиан) определяется как
$$
W(t) = \det \begin{pmatrix}
\mathbf{x}_1(t) & \mathbf{x}_2(t) & \dots & \mathbf{x}_n(t)
\end{pmatrix} = \det \begin{pmatrix}
x_{11}(t) & x_{21}(t) & \dots & x_{n1}(t) \\
x_{12}(t) & x_{22}(t) & \dots & x_{n2}(t) \\
\vdots & \vdots & \ddots & \vdots \\
x_{1n}(t) & x_{2n}(t) & \dots & x_{nn}(t)
\end{pmatrix}
$$


**Т ( Критерий линейной независимости решений ) $:$** если $\mathbf{x}_1(t), \mathbf{x}_2(t), \dots, \mathbf{x}_n(t)$ — решения системы $\mathbf{x}' = A(t)\mathbf{x}$, то $:$
- либо $W(t) \equiv 0$ (и решения линейно зависимы),
- либо $W(t) \neq 0$ для всех $t \in I$ (и решения линейно независимы).

**Доказательство $:$** пусть $W(t_0) = 0$ в некоторой точке $t_0$. Тогда существует нетривиальная линейная комбинация $C_1 \mathbf{x}_1(t_0) + \dots + C_n \mathbf{x}_n(t_0) = \mathbf{0}$. Рассмотрим функцию
$$
\mathbf{y}(t) = C_1 \mathbf{x}_1(t) + \dots + C_n \mathbf{x}_n(t).
$$
Она является решением системы и удовлетворяет начальному условию $\mathbf{y}(t_0) = \mathbf{0}$. По теореме единственности $\mathbf{y}(t) \equiv \mathbf{0}$, значит, $\{\mathbf{x}_i(t)\}$ линейно зависимы, и $W(t) \equiv 0$.

Обратно, если $W(t) \neq 0$ в некоторой точке, то решения линейно независимы.

**Фундаментальная система решений (ФСР)** однородной системы $\mathbf{x}' = A(t)\mathbf{x}$ называется набор из $n$ линейно независимых решений $\{\mathbf{x}_1(t), \dots, \mathbf{x}_n(t)\}$.

**Свойства ФСР $:$**
1. Любое решение системы можно представить в виде линейной комбинации **ФСР** $:$
$$
\mathbf{x}(t) = C_1 \mathbf{x}_1(t) + \dots + C_n \mathbf{x}_n(t).
$$
2. Определитель Вронского для **ФСР** не равен нулю ни в одной точке.

**Критерий ФСР $:$** набор решений является **ФСР** тогда и только тогда, когда их вронскиан $W(t) \neq 0$ хотя бы в одной точке.
