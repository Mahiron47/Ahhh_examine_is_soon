Общее уравнение кривой второго порядка $:$
$$
Ax^2 + 2Bxy + Cy^2 + 2Dx + 2Ey + F = 0\qquad A^2 + B^2 + C^2 \neq 0
$$

Эллипс — кривая второго порядка, геометрическое место точек плоскости, определяемое суммой расстояний от произвольной точки до каждой из точек фокуса эллипса.

Уравнение эллипса $:$
$$
\sqrt{y^2 + (x + c)^2} + \sqrt{y^2 + (x - c)^2} = 2a
$$

После преобразований $:$
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1\qquad b^2 = a^2 - c^2
$$
![Эллипс](../Ресурсы/Ellips.svg)  
Эксцентриситет $-$ характеристика кривой второго порядка, для эллипса $\varepsilon<0$ $:$
$$
\varepsilon = \frac{c}{a} \Big|_{a > b}\qquad \varepsilon = \frac{c}{b} \Big|_{b > a}
$$

Директриса $-$ характеристика кривой второго порядка $:$
$$
x = \pm\frac{a}{\varepsilon} \Big|_{a > b}\qquad y = \pm\frac{b}{\varepsilon} \Big|_{b > a}
$$

### Гипербола

Гипербола — кривая второго порядка, геометрическое место точек плоскости, определяемое абсолютной величиной разности расстояний от произвольной точки до каждой из точек фокуса гиперболы.

Уравнение гиперболы $:$
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1\qquad b^2 = c^2 - a^2
$$

Ассимптоты $:$
$$
x = \pm\frac{a}{b}y\qquad y = \pm\frac{b}{a}x
$$

### Парабола

Парабола — кривая второго порядка, геометрическое место точек, определяемое равенством расстояния от произвольной точки до фокуса и директрисы.

Каноническое уравнение $:$
$$
x^2 = 2py\qquad y^2 = 2px
$$

### Инварианты кривых второго порядка

Инварианты $:$
$$
I_1 = A + C\qquad I_2 = \left|\begin{matrix}A & B \\ B & C\end{matrix}\right|\qquad I_3 = \left|\begin{matrix}A & B & D \\ B & C & E \\ D & E & F\end{matrix}\right|
$$

Классификация кривых по значениям инвариантов $:$

| $\forall I_1$  | $I_3 > 0$   | $I_3 = 0$                     | $I_3 < 0$       |
| -------        | ----------- | ----------------------------- | --------------- |
| $I_2 > 0$      | Эллипс      | Точка                         | Мнимый эллипс   |
| $I_2 = 0$      | Парабола    | Пересекающиеся прямые (две)   | Парабола        |
| $I_2 < 0$      | Гипербола   | Пересекающиеся прямые (две)   | Гипербола       |

### Приведение к каноническому виду

1. **Параллельный перенос** $:$
   - Сгруппировать переменные в полные квадраты.

2. **Поворот системы координат** $:$
   - Найти собственные значения и векторы многочлена.
   - Подставить в уравнение:
     $$
     \lambda_1x^2 + \lambda_2y^2 + 2Dx + 2Ey + F = 0
     $$
   - Построить новую систему координат по собственным векторам.
